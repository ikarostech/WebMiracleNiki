```python
import json
import csv
import pandas as pd
import codecs as cd
```

```python
csv_path = 'MiracleNikkiJp_items.csv'
csv_data = pd.read_csv(csv_path,encoding='utf-8')
```

```python
csv_data.head()
```

```python
colors = []
tags = []
for index,row in csv_data.iterrows():
    c_list = []
    t_list = []
    if row['Color1']==row['Color1']:
        c_list.append(row['Color1'])
    if row['Color2']==row['Color2']:
        c_list.append(row['Color2'])
    if row['tag1']==row['tag1']:
        t_list.append(row['tag1'])
    if row['tag2']==row['tag2']:
        t_list.append(row['tag2'])
    colors.append(c_list)
    tags.append(t_list)
```

```python
print(colors)
```

```python
csv_data['Color'] = colors
csv_data['Tag'] = tags
```

```python
json_data = csv_data
```

```python
json_data = json_data.drop(['SN','Color1','Color2','tag1','tag2','-'],axis=1)
```

```python
json_data.head()
```

```python
json_data = json_data[json_data['Name']==json_data['Name']]
```

```python
json_str = json_data.to_json(orient='records')
print(json_str)
```

```python
json_path = 'MiracleNikkiJp_items.json'

with open(json_path,mode='w') as f:
    f.write(json_str)
    f.close()
```
